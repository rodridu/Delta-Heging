```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (replace 'your_dataset.csv' with your actual file path)
data = pd.read_csv('C:/Users/my/Downloads/a.csv')

# Specify the desired time points
start_time = "09:30"
end_time = "16:30"
time_interval = pd.Timedelta(minutes=15)

# Convert 'quote_datetime' column to datetime format
data['quote_datetime'] = pd.to_datetime(data['quote_datetime'])

# Filter the data for the desired time range
desired_data = data[
    (data['quote_datetime'].dt.time >= pd.to_datetime(start_time).time()) &
    (data['quote_datetime'].dt.time <= pd.to_datetime(end_time).time())
]

# Calculate the days to maturity
desired_data['expiration'] = pd.to_datetime(desired_data['expiration'])
desired_data['days_to_maturity'] = (desired_data['expiration'] - desired_data['quote_datetime']).dt.days

# Initialize lists to store time intervals and corresponding option gains
time_intervals = []
option_gains = []

# Iterate over the time intervals and calculate option gains
current_time = pd.to_datetime(start_time)
while current_time <= pd.to_datetime(end_time):
    interval_data = desired_data[
        (desired_data['quote_datetime'].dt.time >= current_time.time()) &
        (desired_data['quote_datetime'].dt.time < (current_time + time_interval).time())
    ]
    if not interval_data.empty:
        # Calculate the discrete delta-hedged call option gain for the interval
        option_gain = (interval_data.iloc[-1]['close'] - interval_data.iloc[0]['close']) \
                      - (interval_data['delta'] * (interval_data['active_underlying_price'].shift(-1) - interval_data['active_underlying_price'])).sum() \
                      - ((interval_data['days_to_maturity'].diff() / 365) * (interval_data['close'] - interval_data['delta'] * interval_data['active_underlying_price'])).sum()
        
        time_intervals.append(current_time)
        option_gains.append(option_gain)
    
    current_time += time_interval
    
    if (current_time.time() == pd.to_datetime(end_time).time()) or \
       (current_time.time() > pd.to_datetime(end_time).time()):
        # Plot the return graph for the current time interval
        plt.figure(figsize=(10, 6))
        plt.plot(time_intervals, option_gains, marker='o', linestyle='-', color='b')
        plt.xlabel('Time')
        plt.ylabel('Option Gain')
        plt.title(f'Discrete Delta-Hedged Call Option Gains ({current_time.time()} - {current_time + time_interval})')
        plt.xticks(rotation=45)
        plt.grid(True)
        plt.tight_layout()
        plt.show()
        
        # Reset lists for the next time interval
        time_intervals = []
        option_gains = []
```

```python
import pandas as pd

# Load the dataset (replace 'your_dataset.csv' with your actual file path)
data = pd.read_csv('C:/Users/my/Downloads/b.csv')

# Specify the desired time period [t, t+tau] (in minutes)
tau = 15

# Specify the desired time point
desired_time = "9:45"

# Convert 'quote_datetime' column to datetime format
data['quote_datetime'] = pd.to_datetime(data['quote_datetime'])

# Filter the data for the desired time period
desired_data = data[
    (data['quote_datetime'].dt.time >= pd.to_datetime(desired_time).time()) &
    (data['quote_datetime'].dt.time < (pd.to_datetime(desired_time) + pd.Timedelta(minutes=tau)).time())
]

# Calculate the days to maturity
desired_data['expiration'] = pd.to_datetime(desired_data['expiration'])
desired_data['days_to_maturity'] = (desired_data['expiration'] - desired_data['quote_datetime']).dt.days

# Select ATM options with approximately 30 days to maturity
atm_options = desired_data[(desired_data['strike'] <= desired_data['active_underlying_price']) &
                           (desired_data['days_to_maturity'] >= 28) &
                           (desired_data['days_to_maturity'] <= 32)]

# Calculate the discrete delta-hedged call option gains for each day
option_gains = []

for day in atm_options['quote_datetime'].dt.date.unique():
    daily_options = atm_options[atm_options['quote_datetime'].dt.date == day]
    option_gain = (daily_options.iloc[-1]['close'] - daily_options.iloc[0]['close']) \
                  - (daily_options['delta'] * (daily_options['active_underlying_price'].shift(-1) - daily_options['active_underlying_price'])).sum() \
                  - ((daily_options['days_to_maturity'].diff() / 365) * (daily_options['close'] - daily_options['delta'] * daily_options['active_underlying_price'])).sum()
    option_gains.append(option_gain)

option_gain_series = pd.Series(option_gains, index=atm_options['quote_datetime'].dt.date.unique())

print("Discrete Delta-Hedged Call Option Gains:")
print(option_gain_series)
```
