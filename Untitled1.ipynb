```python
import os
import pandas as pd

def merge_csv_files_in_folder(folder_path):
    # Initialize an empty DataFrame to hold the merged data
    merged_data = pd.DataFrame()

    # Loop through each file in the folder
    for filename in os.listdir(folder_path):
        filepath = os.path.join(folder_path, filename)

        # Check if the file is a CSV file
        if filename.endswith('.csv'):
            # Read the CSV file into a DataFrame
            df = pd.read_csv(filepath)

            # Append the DataFrame to the merged_data
            merged_data = merged_data.append(df, ignore_index=True)

    return merged_data

def merge_csv_files_in_folders(root_folder):
    # Initialize an empty DataFrame to hold the merged data
    merged_data = pd.DataFrame()

    # Loop through each folder in the root folder
    for foldername in os.listdir(root_folder):
        folderpath = os.path.join(root_folder, foldername)

        # Check if it is a directory (folder)
        if os.path.isdir(folderpath):
            # Merge CSV files in the folder
            folder_merged_data = merge_csv_files_in_folder(folderpath)

            # Append the folder_merged_data to the overall merged_data
            merged_data = merged_data.append(folder_merged_data, ignore_index=True)

    return merged_data

# Replace 'your_root_folder_path' with the path to the main folder containing subfolders with CSV files
# Example: '/path/to/main_folder'
your_root_folder_path = 'C:/Users/my/Downloads/extracted_data/'
merged_df = merge_csv_files_in_folders(your_root_folder_path)

# Save the merged DataFrame to a new CSV file
# Replace 'output_merged.csv' with your desired output file name
merged_df.to_csv('D:/merged_data.csv', index=False)
```

```python
import os
import dask
import dask.dataframe as dd

@dask.delayed
def read_csv_files(folder_path):
    print("Processing folder:", folder_path)
    dfs = []
    for filename in os.listdir(folder_path):
        filepath = os.path.join(folder_path, filename)
        if filename.endswith('.csv'):
            df = dd.read_csv(filepath)
            dfs.append(df)
    return dfs

def merge_csv_files_in_folders(root_folder):
    print("Merging CSV files in subfolders of:", root_folder)
    delayed_results = []
    for foldername in os.listdir(root_folder): 
        folderpath = os.path.join(root_folder, foldername)
        if os.path.isdir(folderpath):
            delayed_results.append(read_csv_files(folderpath))

    # Execute delayed tasks to get the list of lists of DataFrames
    dfs_list_list = dask.compute(*delayed_results)

    # Flatten the list of lists into a single list of DataFrames
    dfs_list = [df for dfs in dfs_list_list for df in dfs]

    # Concatenate the DataFrames using Dask's `concat` function
    merged_data = dd.concat(dfs_list)

    return merged_data

# Replace 'your_root_folder_path' with the path to the main folder containing subfolders with CSV files
# Example: '/path/to/main_folder'
your_root_folder_path = 'C:/Users/my/Downloads/ET/'
merged_df = merge_csv_files_in_folders(your_root_folder_path)

# Save the merged DataFrame to a new CSV file
# Replace 'output_merged.csv' with your desired output file name
merged_df.to_csv('D:/merged_data.csv', index=False)

# Compute the result (execute the Dask computation and store the result in memory as a pandas DataFrame)
merged_df = merged_df.compute()
```

```python
import matplotlib.pyplot as plt
# Step 3: Time-Series Analysis
df['quote_datetime'] = pd.to_datetime(df['quote_datetime'])  # Convert to datetime data type
df.set_index('quote_datetime', inplace=True)  # Set the datetime column as the DataFrame index

# Step 4: Group by 15-Minute Intervals
# Resample the data to 15-minute intervals and calculate the mean delta value for each interval
delta_by_15min = df.resample('15T')['delta'].mean()

# Step 5: Calculate Delta Changes
delta_changes = delta_by_15min.diff()

# Step 6: Visualization
plt.figure(figsize=(12, 6))
plt.plot(delta_by_15min.index, delta_by_15min.values, label='Mean Delta Value')
plt.xlabel('Time')
plt.ylabel('Delta Value')
plt.title('Mean Delta Value of Options Over 15-Minute Intervals')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.bar(delta_changes.index, delta_changes.values, label='Delta Value Changes')
plt.xlabel('Time')
plt.ylabel('Delta Value Changes')
plt.title('Changes in Mean Delta Value of Options Over 15-Minute Intervals')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 2: Delta Level Classification
# Define delta level categories based on the 'delta' column
low_delta_options = df[df['delta'] < 0.3]
medium_delta_options = df[(df['delta'] >= 0.3) & (df['delta'] < 0.7)]
high_delta_options = df[df['delta'] >= 0.7]

# Step 3: Calculate Option's Risk Measures
# Calculate risk measures for each category of options
low_delta_gamma = low_delta_options['gamma'].mean()
low_delta_theta = low_delta_options['theta'].mean()
low_delta_vega = low_delta_options['vega'].mean()
low_delta_rho = low_delta_options['rho'].mean()

medium_delta_gamma = medium_delta_options['gamma'].mean()
medium_delta_theta = medium_delta_options['theta'].mean()
medium_delta_vega = medium_delta_options['vega'].mean()
medium_delta_rho = medium_delta_options['rho'].mean()

high_delta_gamma = high_delta_options['gamma'].mean()
high_delta_theta = high_delta_options['theta'].mean()
high_delta_vega = high_delta_options['vega'].mean()
high_delta_rho = high_delta_options['rho'].mean()

# Step 4: Analyze Risk Exposure for Different Delta Levels
# Print or plot the calculated risk measures for each delta category
print("Low Delta Options - Gamma:", low_delta_gamma)
print("Low Delta Options - Theta:", low_delta_theta)
print("Low Delta Options - Vega:", low_delta_vega)
print("Low Delta Options - Rho:", low_delta_rho)

print("Medium Delta Options - Gamma:", medium_delta_gamma)
print("Medium Delta Options - Theta:", medium_delta_theta)
print("Medium Delta Options - Vega:", medium_delta_vega)
print("Medium Delta Options - Rho:", medium_delta_rho)

print("High Delta Options - Gamma:", high_delta_gamma)
print("High Delta Options - Theta:", high_delta_theta)
print("High Delta Options - Vega:", high_delta_vega)
print("High Delta Options - Rho:", high_delta_rho)

# Step 5: Visualization (optional)
# Create plots to visualize the risk measures for different delta categories
# Example: bar charts comparing risk measures for different delta categories
delta_categories = ['Low Delta', 'Medium Delta', 'High Delta']
risk_measures = ['Gamma', 'Theta', 'Vega', 'Rho']
data = [[low_delta_gamma, medium_delta_gamma, high_delta_gamma],
        [low_delta_theta, medium_delta_theta, high_delta_theta],
        [low_delta_vega, medium_delta_vega, high_delta_vega],
        [low_delta_rho, medium_delta_rho, high_delta_rho]]

fig, ax = plt.subplots()
for i, risk_measure in enumerate(risk_measures):
    ax.bar(delta_categories, data[i], label=risk_measure)

ax.set_ylabel('Average Risk Measure')
ax.set_xlabel('Delta Level')
ax.legend()
plt.show()
```
