```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (replace 'your_dataset.csv' with your actual file path)
data = pd.read_csv('C:/Users/my/Downloads/b.csv')

# Specify the desired time period [t, t+tau] (in minutes)
tau = 15

# Specify the desired time point
desired_time = "9:30"

# Convert 'quote_datetime' column to datetime format
data['quote_datetime'] = pd.to_datetime(data['quote_datetime'])

# Filter the data for the desired time period
desired_data = data[
    (data['quote_datetime'].dt.time >= pd.to_datetime(desired_time).time()) &
    (data['quote_datetime'].dt.time < (pd.to_datetime(desired_time) + pd.Timedelta(minutes=tau)).time())
]

# Calculate the days to maturity
desired_data['expiration'] = pd.to_datetime(desired_data['expiration'])
desired_data['days_to_maturity'] = (desired_data['expiration'] - desired_data['quote_datetime']).dt.days

# Select ATM options with approximately 30 days to maturity
atm_options = desired_data[(desired_data['strike'] <= desired_data['active_underlying_price']) &
                           (desired_data['days_to_maturity'] >= 28) &
                           (desired_data['days_to_maturity'] <= 32)]

# Calculate the discrete delta-hedged call option gains for each day
option_gains = []

for day in atm_options['quote_datetime'].dt.date.unique():
    daily_options = atm_options[atm_options['quote_datetime'].dt.date == day]
    option_gain = (daily_options.iloc[-1]['close'] - daily_options.iloc[0]['close']) \
                  - (daily_options['delta'] * (daily_options['active_underlying_price'].shift(-1) - daily_options['active_underlying_price'])).sum() \
                  - ((daily_options['days_to_maturity'].diff() / 365) * (daily_options['close'] - daily_options['delta'] * daily_options['active_underlying_price'])).sum()
    option_gains.append(option_gain)

option_gain_series = pd.Series(option_gains, index=atm_options['quote_datetime'].dt.date.unique())

print("Discrete Delta-Hedged Call Option Gains:")
print(option_gain_series)

plt.figure(figsize=(10, 6))
option_gain_series.plot(marker='o', linestyle='-', color='b')
plt.xlabel('Date')
plt.ylabel('Option Gain')
plt.title(f'Discrete Delta-Hedged Call Option Gains at ({desired_time})')
plt.grid(True)
plt.tight_layout()
plt.show()
```
